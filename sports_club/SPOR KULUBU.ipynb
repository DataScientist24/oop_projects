```python
class File:
    def sporcuları_kaydet(self):
        """tüm sporcuları txt dosyasına ekleme metodu"""
        with open("sporcular.txt", "w") as f:
            for j in self.sporcular:
                for sporcu in self.sporcular[j]:
                    f.write(f"{sporcu.kod},{sporcu.get_tcno()},{sporcu.first_name} {sporcu.last_name},{sporcu.get_gmail()}\n")

    def sporcuları_oku(self):
        """sporcuların bilgilerini okuma metodu"""
        mertebe=1
        with open("sporcular.txt", "r") as f:
            for i in f.readlines():
                print(f"dosyada kaydadilmiş {mertebe}. :{i}")
                mertebe+=1
```

spor kulubun yöneticisinin sınıfı 

```python
class SporKulubuYoneticisi:
    """Spor kulüplerini yönetmek için sınıf"""

    def __init__(self,name):
        self.kulupler = []
        self.name=name
    def kulup_ekle(self,kulup,sporKulubu):
        """Kulüp listesine kulüp ekleme yöntemi"""
        sporKulubu.sporcular[kulup.tur]=[]  #[ ] içine sporcularıları ekleyeceğiz
        self.kulupler.append(kulup)

    def kulup_sil(self, kulup):
        """Bir kulübü kulüp listesinden çıkarma yöntemi"""
        self.kulupler.remove(kulup)

    def kulupleri_getir(self):
        """Tüm kulüplerin bir listesini alma yöntemi"""
        return self.kulupler
   
    def kabulmu(self,sporcu,spor):
        while True:
            karar=input(f"{sporcu.first_name} için SporKulubuYoneticisi kararı: ")
            if karar=="True":
                return True
            elif karar=="False":
                return False
    def greet(self):
        print("Hello in executive class") #yönetici sınıfında merhaba demek
```

Hashing bir veri yapısına ekleme ve silme işlemi gerçekleştirmek için bir sınıf oluşturdum

```python
#hashing veri yapısının türüdür
class Hashing:
    def __init__(self,size):
        self.hash_table=[None]*size
        self.size=size
    def hash_function(self,word):
        toplam=0
        for i in range(len(word)):
            toplam+=ord(word[i])
        return toplam%self.size
    def add_hash_table(self,sporcu):
        sporcu_name=sporcu.get_name()
        hash_value=self.hash_function(sporcu_name)
        index=hash_value
        aramak=0
        while aramak<self.size:
            if self.hash_table[index] is None:
                self.hash_table[index]=sporcu
                return 
            else:
                index=(index+1)%self.size
            aramak+=1
        return "yer bulunmadı"

    def Aramak(self,sporcu):
        sporcu_name=sporcu.get_name()
        hash_value=self.hash_function(sporcu_name)
        index=hash_value
        aramak=0
        while aramak<self.size:
            if self.hash_table[index]==sporcu:
                return index
            else:
                index=(index+1)%self.size
            aramak+=1
        return "yer bulunmadı"
    def silmek(self,sporcu):
        index=self.aramak(sporcu)
        if type(index) is not int:
            return "sporcu bulunmadı"
        else:
            self.hash_table[index]=None
    def get_hash_table(self):
        return self.hash_table
    

```
